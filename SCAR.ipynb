```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x_lab_pos, sd_lab_pos = 0, 1.5
x_unlab_pos, sd_unlab_pos = 1.5, 1
x_neg, sd_neg = 4, 1
y, y_sd = 1, 1
lab_pos = np.random.normal(x_lab_pos, sd_lab_pos, 50)
unk_pos = np.random.normal(x_unlab_pos, sd_unlab_pos, 50)
neg =  np.random.normal(x_neg, sd_neg, 100)
y_dist = np.random.normal(y, y_sd, 200)
```

```python
y, y_sd = 1, 1
```

```python
lab_pos = np.random.normal(x_lab_pos, sd_lab_pos, 50)
unk_pos = np.random.normal(x_unlab_pos, sd_unlab_pos, 50)
neg =  np.random.normal(x_neg, sd_neg, 100)
y_dist = np.random.normal(y, y_sd, 200)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(20,4),gridspec_kw={'height_ratios': [2, 1]}, sharex=True)

for i, (pos_color, unk_color) in enumerate(zip(["green","green","white"],["blue","orange","blue"])):
    axs[0,i].text(x=-3.8,y=2.5,s=["A","B","C"][i], fontsize=26)

    axs[0,i].text(x=-3.5,y=1.4,s= "Labeled Positive", color=pos_color, fontsize=14)
    axs[0,i].text(x=0.5,y=-0.9,s= "Unlabeled Positive", color=unk_color, fontsize=14)
    axs[0,i].text(x=4,y=-0.7,s= "Negative", color="orange", fontsize=14)
    axs[0,i].scatter(lab_pos, y_dist[:50],s=50, color=pos_color, marker="+")
    axs[0,i].scatter(unk_pos, y_dist[50:100],s=30, color=unk_color, marker="x")
    axs[0,i].scatter(neg, y_dist[100:],s=40, color="orange", marker="_")

    sns.kdeplot(lab_pos, color=pos_color, ax=axs[1,i])
    if i == 1:
        sns.kdeplot(np.concatenate((unk_pos, neg)), color="orange", ax=axs[1,i])
        axs[0,i].axline((0.5, 1), slope=2, color="grey", linestyle="--")
    else:
        sns.kdeplot(unk_pos, color=unk_color, ax=axs[1,i])
        sns.kdeplot(neg, color="orange", ax=axs[1,i])

    if i == 2:
        axs[0,i].axline((0.5, 1), slope=2, color="lightgrey", linestyle="--")
        axs[0,i].axline((3, 1), slope=-10, color="grey", linestyle="--")
    for ax in axs.transpose()[i]:
        ax.axes.yaxis.set_ticks([])
        ax.set_xlim((-4, 7))
    axs[1,i].set_ylabel("")
    axs[1,i].set_xlabel("Arbitrary Units")


#plt.imshow()
fig.tight_layout()
plt.subplots_adjust(wspace=0.02, hspace=0.1)
plt.savefig("scar_assumption.pdf")
```
